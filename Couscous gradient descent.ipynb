```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
def f(x):
    # monkey saddle
    return x[0] ** 3 - 3 * x[0] * x[1] ** 2

def df(x):
    return np.array([3 * x[0] ** 2 - 3 * x[1] ** 2, 
                     -6 * x[0] * x[1]])
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

fig = plt.figure()
ax = fig.gca(projection="3d")
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
Z = np.hstack((X.reshape(-1, 1), Y.reshape(-1, 1)))

ax.plot_surface(X, Y, np.array([f(z) for z in Z]).reshape(X.shape), 
                rstride=1, cstride=1, cmap=cm.viridis, linewidth=0, antialiased=False)
plt.show()
```

```python
# Couscous gradient descent
x0 = np.random.rand(2)

for i in range(20000):
    if i % 500 == 0:
        print(x0)
    x0 -= 0.05 * (2 * np.random.rand() - 1) * df(x0) 
```
